```python
r = requests.get("https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
content = r.content
soup=BeautifulSoup(content, "html.parser")
soup.prettify()
temp = soup.find_all('div', { 'class': 'propertyRow' })
temp[0].find('h4', {"class":"propPrice"}).text.strip()
l=[]
for x in temp:
    d = {}
    d['price'] = x.find('h4', {"class":"propPrice"}).text.strip()
    d['address'] = x.find_all('span', {"class":"propAddressCollapse"})[0].text.strip()
    d['locality'] = x.find_all('span', {"class":"propAddressCollapse"})[1].text.strip()
    try:
        d['beds'] = x.find('span', {"class":"infoBed"}).find('b').text.strip()
    except:
        d['beds'] = 0
    try:
        d['area'] = x.find('span', {"class":"infoSqFt"}).find('b').text.strip()
    except:
        d['area'] = None
    try:
        d['full_baths'] = x.find('span', {"class":"infoValueFullBath"}).find('b').text.strip()
    except:
        d['full_baths'] = 0
    try:
        d['half_baths'] = x.find('span', {"class":"infoValueHalfBath"}).find('b').text.strip()
    except:
        d['half_baths'] = 0
    for column_group in x.find_all('div', {'class':'columnGroup'}):
        for group, name in zip(column_group.find_all('span', {'class':'featureGroup'}), column_group.find_all('span', {'class':'featureName'})):
            d[group.text.strip()[:-1]] = name.text
    l.append(d)
    
print(l)
```

```python
df=pandas.DataFrame(l)
df.to_csv('data.csv')
df
```
